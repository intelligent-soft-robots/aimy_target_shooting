## Orientation Actuators

```python
actuation = [i * 0.1 for i in range(0, 11)]
actuation = np.array(actuation)

xloc = plticker.MultipleLocator(base=0.2)
yloc_phi = plticker.MultipleLocator(base=5)
yloc_theta = plticker.MultipleLocator(base=3)
```

```python
# Phi
phi = [21.0, 18.5, 15.0, 10.0, 5.0, 0.0, -5.0, -10.0, -15.0, -17.0, -19.5]
phi_pos = [
    16.81766552,
    12.41133775,
    4.05559336,
    0.0,
    -4.26665414,
    -11.6580105,
    -16.54297629,
]
phi_pos_actuation = [i * (1 / 6) for i in range(7)]

phi_cad = np.linspace(23, -23, 11)
phi_linear = np.linspace(phi[0], phi[-1], 11)

# Theta
theta = [6.5, 8, 10, 12, 14, 16, 18, 20.5, 23, 25.5, 27.1]
theta_pos = [6.40400874, 12.17836522, 19.90385964, 28.11300675, 37.14585762]

theta_pos_actuation = [i * 0.25 for i in range(5)]
theta_cad = np.linspace(9, 31, 11)
theta_linear = np.linspace(theta[0], theta[-1], 11)

# Function
f_phi = interpolate.interp1d(phi_pos_actuation, phi_pos)
f = interpolate.interp1d(theta_pos_actuation, theta_pos)

# Plotting
fig, axs = plt.subplots(1, 2, figsize=(4, 2), layout="constrained")
# axs[0].plot(actuation, phi_cad, linestyle="dotted", color="#46b361")
axs[0].plot(actuation, f_phi(actuation), label="measured", color="#0193d7")
axs[1].plot(actuation, f(actuation), label=r"measured", color="#CF5369")


for ax in axs:
    ax.legend()
    ax.grid()
    ax.set_xlabel(r"Actuation [0,1]")

axs[0].set_ylabel(r"$\varphi$ [\textdegree]")
axs[1].set_ylabel(r"$\theta$ [\textdegree]")

axs[0].xaxis.set_major_locator(xloc)
axs[0].yaxis.set_major_locator(yloc_phi)

axs[1].xaxis.set_major_locator(xloc)
axs[1].yaxis.set_major_locator(yloc_theta)

fig.savefig("plots/orientation_actuators.pdf")
```
